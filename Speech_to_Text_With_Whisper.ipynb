```python
# Import the libraries
import whisper
import torch
import os

# Set the device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the model
whisper_model = whisper.load_model("large", device=device)
```

```python
# Install the library
!pip -qqq install pytube

# import the module
from pytube import YouTube
```

```python
def video_to_audio(video_URL, destination, final_filename):

  # Get the video
  video = YouTube(video_URL)

  # Convert video to Audio
  audio = video.streams.filter(only_audio=True).first()

  # Save to destination
  output = audio.download(output_path = destination)

  _, ext = os.path.splitext(output)
  new_file = final_filename + '.mp3'

  # Change the name of the file
  os.rename(output, new_file)

# Video to audio
video_URL = 'https://www.youtube.com/watch?v=E9lAeMz1DaM'
destination = "."
final_filename = "motivational_speech"
video_to_audio(video_URL, destination, final_filename)
```

## Start Transcription

#### English Audio

```python
audio_file = "motivational_speech.mp3"
result = whisper_model.transcribe(audio_file)
```

```python
# Imports used through the rest of the notebook.
import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F

import IPython

from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_audio, load_voice, load_voices

# This will download all the models used by Tortoise from the HuggingFace hub.
tts = TextToSpeech()
```

```python
# This is the text that will be spoken.
text = urdu_to_english

# Pick a "preset mode" to determine quality. Options: {"ultra_fast", "fast" (default), "standard", "high_quality"}. See docs in api.py
preset = "fast"
```

```python
# Optionally, upload use your own voice by running the next two cells. I recommend
# you upload at least 2 audio clips. They must be a WAV file, 6-10 seconds long.
CUSTOM_VOICE_NAME = "bilal"

import os
from google.colab import files

custom_voice_folder = f"tortoise/voices/{CUSTOM_VOICE_NAME}"
os.makedirs(custom_voice_folder)
for i, file_data in enumerate(files.upload().values()):
  with open(os.path.join(custom_voice_folder, f'{i}.wav'), 'wb') as f:
    f.write(file_data)
```

```python
# Generate speech with the custotm voice.
voice_samples, conditioning_latents = load_voice(CUSTOM_VOICE_NAME)
gen = tts.tts_with_preset(str(urdu_to_english), voice_samples=voice_samples, conditioning_latents=conditioning_latents,
                          preset=preset)
torchaudio.save(f'generated-{CUSTOM_VOICE_NAME}.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio(f'generated-{CUSTOM_VOICE_NAME}.wav')
```

```python
# Import necessary libraries and modules at the beginning of your script
# import torch
# import your_tts_module  # Replace 'your_tts_module' with the actual module name
# import torchaudio
# import IPython.display



# Define your input text (assuming 'urdu_to_english' contains the text)
input_text = str(urdu_to_english)

# Split the input text into segments (adjust the segment length as needed)
segment_length = 200  # Adjust this based on your model's limit
text_segments = [input_text[i:i+segment_length] for i in range(0, len(input_text), segment_length)]

# Initialize an empty list to store generated audio segments
generated_audio_segments = []

# Load the voice and conditioning latents (replace with actual data)
voice_samples, conditioning_latents = load_voice(CUSTOM_VOICE_NAME)

# Process each text segment separately
for segment in text_segments:
    gen = tts.tts_with_preset(segment, voice_samples=voice_samples, conditioning_latents=conditioning_latents, preset=preset)
    generated_audio_segments.append(gen)

# Concatenate the audio segments if needed
final_audio = torch.cat(generated_audio_segments, dim=-1)

# Save the final audio as a WAV file
torchaudio.save(f'generated-{CUSTOM_VOICE_NAME}.wav', final_audio.squeeze(0).cpu(), 24000)

# Display the generated audio
IPython.display.Audio(f'generated-{CUSTOM_VOICE_NAME}.wav')
```
